# Atomic Selection Rules

In this notebook we will study, computationally, selection rules and transition probabilities for atomic transitions.

## Transition Moments and Selection Rules

Consider an atom or molecule whose wavefunctions can be accurately approximated by Slater determinants, e.g. $\Phi_i = | \phi_{i_1} \phi_{i_2} \ldots \phi_{i_N} |$ for the initial state and $\Phi_f = | \phi_{f_1} \phi_{f_2} \ldots \phi_{f_N} |$ for the final state. Transitions between these determinants are governed by selection rules.

- The first selection rule is that transitions are forbidden if the initial and final wavefunctions differ by more than a single orbital. (I.e., double- and higher-order excitations are forbidden at the Slater determinant level.) Thus allowed transitions are all single-orbital transitions.
- The second selection rule is that transitions between spin-orbitals with different spin are forbidden. In the absence of spin-orbit coupling or magnetic fields, one cannot flip a spin when exciting an electron. (Indeed, this is a sub-case of the more fundamental rule that spin is a preserved quantum number in nonrelativsitic quantum mechanics.)

The transition between two orbitals with the same spin, $\phi_i(\mathbf{r})$ and $\phi_f(\mathbf{r})$ is allowed only if the matrix element of the perturbation operator is nonzero. The simplest of these operators is the dipole operator. For an atom, the square of the transition dipole moment is:

$$
\mu_{if}^2 = \left|\Braket{\phi_i | x | \phi_f}\right|^2 + \left|\Braket{\phi_i | y | \phi_f}\right|^2 + \left|\Braket{\phi_i | z | \phi_f}\right|^2
$$

This is the appropriate operator for an atom that is not oriented; if the system is oriented with respect to the impending light then only the dipole moment directed along the direction of the light's electric field is relevant.

Other types of (weaker) transitions have the same general formula. For example, magnetic dipole transitions are governed by the square of the magnetic transition dipole,

$$
\ell_{if}^2 = \left|\Braket{\phi_i | \hat{L}_x | \phi_f}\right|^2+\left|\Braket{\phi_i | \hat{L}_y | \phi_f}\right|^2+\left|\Braket{\phi_i | \hat{L}_z | \phi_f}\right|^2
$$

and quadrupole transitions are governed by the expression:

$$
\theta_{if}^2 = \left|\Braket{\phi_i | xz | \phi_f}\right|^2 + \left|\Braket{\phi_i | yz | \phi_f}\right|^2 + \left|\Braket{\phi_i | xy | \phi_f}\right|^2+\tfrac{1}{2} \left|\Braket{\phi_i | x^2-y^2 | \phi_f}\right|^2 + \tfrac{1}{6}\left|\Braket{\phi_i | 2z^2 - x^2-y^2 | \phi_f}\right|^2
$$

The atomic selection rules are a direct consequence of the situations where the transition dipole moment vanishes. For example, the electric dipole selection rules for atoms are that

1. the total orbital angular momentum changes by at most one unit, $L_f - L_i = {0,\pm 1}$, but $L_f =0 \rightarrow L_i = 0$ transitions are forbidden. In addition, for the initial and final orbital, one must have $\ell_f - \ell_i = \pm 1$.
2. The total spin does not change, $S_f - S_i = 0$.
3. the total angular momentum changes by at most one unit, $J_f - J_i = {0, \pm 1}$, but $J_f=0 \rightarrow J_i = 0$ transitions are forbidden.

> N.B. Usually the expressions for the transition dipole/quadrupole moment and the magnetic dipole moment include a dependence on the electric charge and Bohr magneton, respectively. I'm using atomic coordinates, but this means I'll need to (re)introduce the electric charge later. This is convenient since the dependence on the electric charge can then be immediate subsumed into the fine structure constant.


## Spontaneous Emission

In the absence of an light and other external perturbations, excited states of molecules still spontaneously decay to lower-energy states. The quantum mechanical process of spontaneous emission results from what are called vacuum oscillations; you can think of this as the zero-point oscillations of the electromagnetic field, or as a manifestation of a Heisenberg uncertainty relation, whereby a photon can be created from the vacuum for a short time, due through a energy-time uncertainty principle. This fleeting photon can stimulate emission of the excited state, and then the photon is reabsorbed by the vacuum. The rate of spontaneous emission from an excited state $f$ to a lower-energy state $i$ is:

$$
\begin{split}
A_{if} &= \omega_{if}^3 \frac{e^2|\mu_{if}|^2}{3 \pi \epsilon_0 \hbar c^3} \\
 &= \frac{4 \alpha}{3c^2} \omega_{if}^3 |\mu_{if}|^2
\end{split}
$$

where

$$
\alpha=\frac{1}{4 \pi \epsilon_0} \frac{e^2}{\hbar c}=7.29735 \times 10^{-3}
$$

is the fine structure constant, and

$$
\omega_{if}=\frac{E_{i}-E_f}{\hbar}
$$

is the angular frequency of the transition. The expression for the rate of quadrupole transitions is similar,

$$
A_{if} =  \frac{4 k^2 \alpha}{3c^2} \omega_{fi}^3 |\theta_{fi}|^2
$$

where $k=\tfrac{2 \pi}{\lambda}$.


## Lifetime Broadening

For an atomic gas at extremely low pressure and temperature, the breadth of the spectral lines is controlled by the lifetime of the atomic states; narrow spectral lines correspond to long-lived states. The spectral lineshape is a Lorentzian,

$$
L(\omega) =  \frac{(2 \pi \tau)^{-1}}{(2 \tau)^{-2}+(\omega - \omega_{if})^2}
$$

Here the lifetime of the state is,

$$
\tau = \frac{1}{A_{if}}
$$

and so $A_{if}$ is the full-width at half-maximum of the spectral line.
